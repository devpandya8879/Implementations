```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel("Book1.xlsx")
```

```python
df.head(50)
```

```python
plt.scatter(df['Rollno'],df['Marks'])
plt.xlabel("rollno")
plt.ylabel("marks")
```

```python
km=KMeans(n_clusters=3)
predicted=km.fit_predict(df[['Rollno','Marks']])
predicted
```

```python
df['clusters']=predicted
```

```python
df.head(50)
```

```python
df1=df[df.clusters==0]
df2=df[df.clusters==1]
df3=df[df.clusters==2]

plt.scatter(df1['Rollno'],df1['Marks'],color="red")
plt.scatter(df2['Rollno'],df2['Marks'],color="blue")
plt.scatter(df3['Rollno'],df3['Marks'],color="green")
```

```python
scale=MinMaxScaler()
scale.fit(df[['Marks']])
df['Marks']=scale.transform(df[['Marks']])

scale.fit(df[['Rollno']])
df['Rollno']=scale.transform(df[['Rollno']])
df
```

```python
km=KMeans(n_clusters=3)
predicted=km.fit_predict(df[['Rollno','Marks']])
predicted
```

```python
df=df.drop(['clusters'],axis='columns')
df['clusters']=predicted
df.head
```

```python
df1=df[df.clusters==0]
df2=df[df.clusters==1]
df3=df[df.clusters==2]

plt.scatter(df1['Rollno'],df1['Marks'],color="red")
plt.scatter(df2['Rollno'],df2['Marks'],color="blue")
plt.scatter(df3['Rollno'],df3['Marks'],color="green")
```
